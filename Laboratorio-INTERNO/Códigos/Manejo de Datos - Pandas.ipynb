```python
# Lambda es una función muda y recibe argumento,
# en este caso es x, pero puede haber más separando con ,

# A la columna de dataframe ugly, que se llama "RA (hh mm ss.sss)"
# se renombrará aplicándole un map (aplica a toda la columna) una
# función local (lambda) que recibe de argumento x, y x se relaciona
# con cada elemento de la columna, a la cual se le va a cambiar de sus
# sus valores los ":" por " ", como lo hicimos anteriormente.
ugly["RA (hh mm ss.sss)"] = ugly["RA (hh mm ss.sss)"].map(lambda x: " ".join(x.split(":")))
```
